THE MOST IMPORTANT RESOURCE IS matplotlib.org/gallery

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,10)
```

```python
plt.plot(x)
plt.show()
```

```python
y = x ** 2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('Number')
plt.ylabel('Square')
plt.title('Squar Curve')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
```

```python
axes = fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
axes.plot(x,y)
axes.set_xlabel('Xlabel')
axes.set_ylabel('Ylabel')
fig
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(y,x,'r')
axes2 = fig.add_axes([0.5,0.22,0.4,0.4])
axes2.plot(x,y)
axes1.set_xlabel('Xlabel')
axes1.set_ylabel('Ylabel')
axes2.set_xlabel('X label')
axes2.set_ylabel('Y label')
```

```python
fig,axes = plt.subplots()
```

```python
fig,axes = plt.subplots()
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=2)
# for removing overlap
plt.tight_layout()
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(y,x,'r')
axes[0].set_title('First Plot')
axes[1].plot(x,y,'g')
axes[1].set_title('Second Plot')
```

```python
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,4)) # we can also use dpi in parameters for pixals per inch 
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(5,3))
ax[0].plot(x,y,'r')
ax[1].plot(y,x,'v')
plt.tight_layout()
```

```python
fig
```

```python
fig.savefig('matplotlib_subplot.png',dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='X squared')
ax.plot(x,x**3,label='X cube')
ax.legend()
```

To set the position of legemd use below put loc value as below
 

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='X squared')
ax.plot(x,x**3,label='X cube')
ax.legend(loc=1)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='X squared')
ax.plot(x,x**3,label='X cube')
ax.legend(loc=(0.1,0.1)) #modified location
```

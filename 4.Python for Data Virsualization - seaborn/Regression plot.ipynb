```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='tip', y='total_bill', data=tips)
```

```python
sns.lmplot(x='tip', y='total_bill', data=tips, hue='smoker', markers=['o','v'])
```

```python
sns.lmplot(x='tip', y='total_bill', data=tips, col='sex')
```

```python
sns.lmplot(x='tip', y='total_bill', data=tips, col='sex', row='time', hue='smoker')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day', row='smoker',hue='sex',palette='coolwarm')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='rainbow',
          aspect=0.8,height=8)
```

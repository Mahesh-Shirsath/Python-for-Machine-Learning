# Finance Data Project 

```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
```

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
```

## Data

We need to get data using pandas datareader. We will get stock information for the following banks:
*  Bank of America
* CitiGroup
* Goldman Sachs
* JPMorgan Chase
* Morgan Stanley
* Wells Fargo

** Figure out how to get the stock data from Jan 1st 2006 to Jan 1st 2016 for each of these banks. Set each bank to be a separate dataframe, with the variable name for that bank being its ticker symbol.**

```python
BAC = data.DataReader("BAC", 'yahoo', start, end) #Bank of America
```

```python
BAC.head()
```

```python
C = data.DataReader("C", 'yahoo', start, end) #CitiGroup
GS = data.DataReader("GS", 'yahoo', start, end) #Goldman Sachs
JPM = data.DataReader("JPM", 'yahoo', start, end) #JPMorgan Chase
MS = data.DataReader("MS", 'yahoo', start, end) #Morgan Stanly
WFC = data.DataReader("WFC", 'yahoo', start, end) #Wells Frago 
```

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

```python
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1, keys=tickers)
```

```python
bank_stocks.head()
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

```python
bank_stocks['BAC']['Close'].max()
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

** Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
BAC['Close'].loc['2008-01-01':'2008-12-31'].rolling(window=30).mean().plot(figsize=(12,4), label='30 day Mov Avg')
BAC['Close'].loc['2008-01-01':'2008-12-31'].plot(figsize=(12,4),label='BAC Close')
plt.legend()
```
